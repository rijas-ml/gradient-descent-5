```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import StandardScaler
```

```python
# (House sizes in sq ft)
X = np.array([600, 800, 1000, 1200, 1500]).reshape(-1, 1)
# (House prices in ₹10000s)
y = np.array([150, 200, 250, 300, 400])
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
model = SGDRegressor(max_iter=1000, learning_rate='constant', eta0=0.01, random_state=42)
model.fit(X_scaled, y)
```

```python
print("Slope (Weight):", model.coef_[0])
print("Intercept (Bias):", model.intercept_[0])
```

```python
y_pred = model.predict(X_scaled)
```

```python
print("Actual prices:", y)
print("Predicted prices:", y_pred)
```

```python
plt.scatter(X, y, color='blue', label='Actual')
plt.plot(X, y_pred, color='red', label='Predicted (GD)')
plt.xlabel('House Size (sq ft)')
plt.ylabel('Price ($1000s)')
plt.title('House Price Prediction using Gradient Descent')
plt.legend()
plt.show()
```
